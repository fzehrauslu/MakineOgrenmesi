```
from sklearn import tree
import keras
from keras.datasets import fashion_mnist
from sklearn.metrics import confusion_matrix
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```
NuOfImages_x_train= x_train.shape[0]
NuOfImages_x_test= x_test.shape[0]
WidthOfImages= x_train.shape[1]
x_train=np.reshape(x_train,(NuOfImages_x_train,WidthOfImages*WidthOfImages))
x_test=np.reshape(x_test,(NuOfImages_x_test,WidthOfImages*WidthOfImages))
```

```
clf_50 = tree.DecisionTreeClassifier(max_depth=50)
clf_50 = clf_50.fit(x_train, y_train)
accuracy_50= clf_50.score(x_test,y_test)
print("accuracy:",accuracy_50)
sns.heatmap(confusion_matrix(y_test, clf_50.predict(x_test)), annot=True)
```

```
clf_30 = tree.DecisionTreeClassifier(max_depth=30)
clf_30 = clf_30.fit(x_train, y_train)
accuracy_30= clf_30.score(x_test,y_test)
print("accuracy:",accuracy_30)
sns.heatmap(confusion_matrix(y_test, clf_30.predict(x_test)), annot=True)
```

```
clf_70 = tree.DecisionTreeClassifier(max_depth=70)
clf_70 = clf_70.fit(x_train, y_train)
accuracy_70= clf_70.score(x_test,y_test)
print("accuracy:",accuracy_70)
#print(confusion_matrix(y_test, clf_70.predict(x_test)))
sns.heatmap(confusion_matrix(y_test, clf_70.predict(x_test)), annot=True)
```

```
clf_50_e = tree.DecisionTreeClassifier(max_depth=50,criterion="entropy")
clf_50_e = clf_50_e.fit(x_train, y_train)
accuracy_50_e= clf_50_e.score(x_test,y_test)
print("accuracy:", accuracy_50_e)
#print(confusion_matrix(y_test, clf_50_e.predict(x_test)))
sns.heatmap(confusion_matrix(y_test, clf_50_e.predict(x_test)), annot=True)
```
